## Seaborn Parameters

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(data=tips, x="total_bill", hue='sex', kde=True)
plt.show() 
```

```python
sns.displot(data=tips, x="total_bill", y='tip', hue='sex')
plt.show() 
```

```python
sns.displot(data=tips, x="total_bill", hue='sex', kind='kde', legend=True, palette='dark')
plt.show() 
```

## Distributions

```python
sns.histplot(data=tips,x='tip',bins=20, cumulative=False, hue='sex', stat='frequency', multiple='dodge', shrink=1)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip',hue='sex',cumulative=False,fill=True,bw_adjust=0.5)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip',hue='sex', kde=True)
plt.show() 
```

## Categorical

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True, color='white')
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box',col='time')
plt.show()
```

## Relation

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day', style='time',size='size')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
plt.show()
```

```python
marker = {'Lunch':'D','Dinner':'o'}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day', style='time',size='size',markers=marker)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day', style='time',size='size')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style="smoker", size="size", kind="scatter", col='time')
plt.show()
```

## Jointplot and Pairplot

```python
sns.jointplot(data=tips,x='total_bill',y='tip', hue='sex', kind='hist', marginal_ticks=True,
              marginal_kws=dict(bins=15, multiple='stack'))
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex')
plt.show()
```

## Heatmap

```python
sns.heatmap(data=tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=5, linecolor='black',
            vmin=-1, vmax=1, center=0, square=True, mask=None, ax=None)
```
