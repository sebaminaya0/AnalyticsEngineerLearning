# Matplotlib for Data Viz

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create data

```python
x = np.linspace(0, 5, 11)
y = x ** 2
w = np.sin(x)
z = np.cos(x)
v = np.tan(x)
country = ['USA', 'Canada', 'Germany', 'UK', 'France']
population = [325, 35, 82, 67, 65]
data = np.random.randint(1,50,100)
```

## Pyplot Basicos

```python
plt.plot(x, y, color='red', linestyle='--')
plt.title('Line Plot')
plt.show()
```

```python
plt.hist(x)
plt.title("Histogram")
plt.show()
```

```python
plt.pie(x)
plt.title("Pie Chart")
plt.show()
```

```python
plt.scatter(x, y, color='red', marker='o', s=30)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.boxplot(x)
plt.title('Boxplot')
plt.show()
```

## Subplot

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x, 'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x, 'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

## Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y, 'r')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

axes.plot(x, y, 'r')
axes2.plot(y, x, 'g')
axes2.set_facecolor('y')
```

## Subplots

```python
fig, axes = plt.subplots()
axes.plot(x, w, 'r--')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, w, 'r--')
axes[1].plot(w, x, 'r--')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x, w, 'r--')
axes[1,1].plot(w, x, 'r--')
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x, w, 'r--')
ax2.plot(w, x, 'r--')
ax3.plot(x, y, 'r--')
ax4.plot(y, x, 'r--')
ax5.plot(x, z, 'r--')
ax6.plot(z, x, 'r--')
ax7.plot(x, v, 'r--')
ax8.plot(v, x, 'r--')
fig.tight_layout()
```

## Legends, labels, titles, and sizes

```python
fig, axes = plt.subplots(1,2,figsize=(10,4))
axes[0].plot(x, y, label='x^2')
axes[0].set_title('X vs Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend(loc=0)
axes[1].plot(y, x)
axes[1].set_title('Y vs X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x, y, label='x^2')
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right', bbox_to_anchor=(1.2, 0.5))
```

```python
# Modify using Matlab syntax

fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
```

```python
# Modify using Pyplot syntax

fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, color="red", alpha=0.5, linestyle="dashed", linewidth=3, marker="o", markersize=10)
ax.plot(x, x+2, color="#1155dd", linestyle="dashdot", linewidth=4)
ax.plot(x, x+3, color="yellow", lw=3, ls="solid", marker="D", markersize=10, markerfacecolor="red")
ax.plot(x, x+4, color="black", lw=5, linestyle="-.")
```

## Bar Plot

```python
plt.bar(country, population, width=0.5, color=['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'purple', 'brown'])
plt.xlabel('Countries')
plt.ylabel('Population')
plt.xticks(country, rotation=45)
plt.title('Population of Countries', fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'green'})
plt.show()
```

```python
plt.barh(country, population, color=['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'purple', 'brown'])
plt.ylabel('Countries')
plt.xlabel('Population')
plt.yticks(country, rotation=45)
plt.title('Population of Countries', fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'green'})
plt.show()
```
