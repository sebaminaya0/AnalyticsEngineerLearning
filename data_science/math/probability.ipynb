# Probability

## What is probability?

Probability is the measure of the likelihood that an event will occur. Probability quantifies as a number between 0 and 1, where, loosely speaking, 0 indicates impossibility and 1 indicates certainty. The higher the probability of an event, the more likely it is that the event will occur.

A simple example is the tossing of a fair (unbiased) coin. Since the coin is fair, the two outcomes ("heads" and "tails") are both equally probable; the probability of "heads" equals the probability of "tails"; and since no other outcomes are possible, the probability of either "heads" or "tails" is 1/2 (which could also be written as 0.5 or 50%).

## Probability Axions

### Axiom 1: Base Axiom

The probability of an event is a number greater than or equal to 0 and less than or equal to 1.

$ 0 \leq P(A) \leq 1 $

### Axiom 2: Impossible Event

If an event cannot occur, its probability is 0.

$ P(\emptyset) = 0 $

### Axiom 3: Certain Event

If an event is certain to occur, its probability is 1.

$ P(S) = 1 $

### Axiom 4: Additive Rule

The probability that one of two or more mutually exclusive events will occur is the sum of the probabilities of the events.

$ P(A \cup B) = P(A) + P(B) $

### Axiom 5: Complement Rule

The probability that an event will not occur is 1 minus the probability that the event will occur.

$ P(A^c) = 1 - P(A) $

### Axiom 6: Multiplicative Rule

The probability that two events both occur is the product of the probabilities of the two events.

$ P(A \cap B) = P(A) \times P(B) $

## Probability Types

### Marginal Probability

Marginal probability is the probability of an event happening without any other conditions. For example, if we have two events A and B, the marginal probability of A is:

$ P(A) $

For example, an event A is rolling a 2 on a dice and an event B is rolling a 3 on a dice. The marginal probability of A is:

$ P(A) = \frac{1}{6} $

### Conditional Probability

Conditional probability is the probability of an event happening given some other event has already happened. For example, if we have two events A and B, the conditional probability of A given B is:

$ P(A|B) $

For example, an event A is rolling a 2 on a dice and an event B is rolling a 3 on a dice. The conditional probability of A given B is:

$ P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{\frac{1}{36}}{\frac{1}{6}} = \frac{1}{6} $

### Joint Probability

Joint probability is the likelihood of two independent events happening at the same time. For example, if we have two events A and B, the joint probability of A and B is:

$ P(A \cap B) $

For example, an event A is rolling a 2 on a dice and an event B is rolling a 3 on a dice. The joint probability of A and B is:

$ P(A \cap B) = P(A) \times P(B) = \frac{1}{6} \times \frac{1}{6} = \frac{1}{36} $

The joint probability based on the marginal and conditional probability is:

$ P(A \cap B) = P(A) \times P(B|A) = P(B) \times P(A|B) $

## Probability Distributions

### Discrete Probability Distribution

A discrete probability distribution is a probability distribution characterized by a probability mass function. A probability mass function is a function that gives the probability that a discrete random variable is exactly equal to some value.

### Continuous Probability Distribution

A continuous probability distribution is a probability distribution characterized by a probability density function. A probability density function is a function that gives the probability that a continuous random variable is exactly equal to some value.

## How to Calculate Probability Distribution

### Discrete Probability Distribution

To calculate the probability distribution of a discrete random variable, we can use the following formula:

$ P(X = x) = \sum_{i=1}^{n} P(X = x_i) $

Where:

- $ P(X = x) $ is the probability of the discrete random variable X being equal to x.

- $ x $ is the value of the discrete random variable X.

- $ P(X = x_i) $ is the probability of the discrete random variable X being equal to $ x_i $.

- $ n $ is the number of possible values of the discrete random variable X.

### Continuous Probability Distribution

To calculate the probability distribution of a continuous random variable, we can use the following formula:

$ P(a \leq X \leq b) = \int_{a}^{b} f(x) dx $

Where:

- $ P(a \leq X \leq b) $ is the probability of the continuous random variable X being between a and b.

- $ a $ is the lower bound of the range.

- $ b $ is the upper bound of the range.

- $ f(x) $ is the probability density function of the continuous random variable X.

## Types of Probability Distributions

### Bernoulli Distribution

The Bernoulli distribution is a discrete probability distribution of a random variable which takes the value 1 with probability p and the value 0 with probability q = 1 - p, where p is a parameter of the distribution.

### Binomial Distribution

Having n independent trials, each of which has a probability p of success, the probability that there are exactly k successes is given by the probability mass function:

$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $

Where:

- $ P(X = k) $ is the probability of k successes.

- $ n $ is the number of trials.

- $ k $ is the number of successes.

- $ p $ is the probability of success.

![Binomial Distribution](https://upload.wikimedia.org/wikipedia/commons/7/75/Binomial_distribution_pmf.svg)

### Multinomial Distribution

The multinomial distribution is a generalization of the binomial distribution. It is used when there are more than two possible outcomes. For example, if we have a dice with 6 sides, the multinomial distribution can be used to calculate the probability of rolling a 1, 2, 3, 4, 5, or 6.

$ P(X_1 = x_1, X_2 = x_2, ..., X_k = x_k) = \frac{n!}{x_1! x_2! ... x_k!} p_1^{x_1} p_2^{x_2} ... p_k^{x_k} $

Where:

- $ P(X_1 = x_1, X_2 = x_2, ..., X_k = x_k) $ is the probability of $ x_1 $ successes for the first outcome, $ x_2 $ successes for the second outcome, ..., and $ x_k $ successes for the kth outcome.

- $ n $ is the number of trials.

- $ x_1 $ is the number of successes for the first outcome.

- $ x_2 $ is the number of successes for the second outcome.

- $ x_k $ is the number of successes for the kth outcome.

- $ p_1 $ is the probability of success for the first outcome.

- $ p_2 $ is the probability of success for the second outcome.

- $ p_k $ is the probability of success for the kth outcome.

### Other distributions

There are many other distributions such as the Poisson distribution, the geometric distribution, the negative binomial distribution, the hypergeometric distribution, the exponential distribution, the gamma distribution, the beta distribution, the chi-squared distribution, the F-distribution, the t-distribution, the Cauchy distribution, the Weibull distribution, the Pareto distribution, etc.

## Let's Code: Binomial Distribution

```python
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt
```

### Theorical Binomial Distribution

$ P(k, n; p) = \binom{n}{k} p^k (1 - p)^{n - k} = \frac{n!}{k! (n - k)!} p^k (1 - p)^{n - k} $

Where:

- $ P(X = k) $ is the probability of k successes.

- $ n $ is the number of trials.

- $ k $ is the number of successes.

- $ p $ is the probability of success.

```python
# Let's build a binomial distribution

def binomial_pmf(n, k, p):
    return factorial(n) / (factorial(k) * factorial(n - k)) * p**k * (1 - p)**(n - k)

def binomial_cdf(n, k, p):
    return sum([binomial_pmf(n, i, p) for i in range(k + 1)])

binomial_pmf(3, 2, 0.5), binomial_cdf(3, 2, 0.5)
```

```python
# Let's test our function against scipy's

dist = binom(5, 0.5)
dist.pmf(3), dist.cdf(3) # Success!
```

```python
# Let's simulate a binomial distribution

p = 0.5
n = 3
binomial(n, p)
```

```python
# Let's simulate multiple binomial distributions

arr = [binomial(n, p) for i in range(1000)]

# Let's plot the results

def plot_hist(trials):
    values = [0, 1, 2, 3]
    arr = [binomial(n, p) for i in range(trials)]
    sim = np.unique(arr, return_counts=True)[1] / trials
    the = [binom(n, p).pmf(i) for i in values]
    plt.bar(values, sim, color='red')
    plt.bar(values, the, color='blue', alpha=0.5)
    plt.legend(['simulation', 'theory'])
    plt.title(f'Binomial Distribution Simulation for {trials} trials (n={n}, p={p})')
    plt.show()

plot_hist(1000)
```

## Let's Code: Normal Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Theorical Normal Distribution

$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2} (\frac{x - \mu}{\sigma})^2} $

Where:

- $ f(x) $ is the probability density function.

- $ \sigma $ is the standard deviation.

- $ \mu $ is the mean.

```python
def gaussian(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * (x - mu)**2 / sigma**2)

def plot_gaussian_pmf(mu, sigma):
    x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
    y = gaussian(x, mu, sigma)
    plt.plot(x, y)
    plt.title(f'Gaussian Distribution (mu={mu}, sigma={sigma})')
    plt.show()

def plot_gaussian_cdf(mu, sigma):
    x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
    y = norm.cdf(x, mu, sigma)
    plt.plot(x, y)
    plt.title(f'Gaussian Distribution (mu={mu}, sigma={sigma})')
    plt.show()

plot_gaussian_pmf(0, 1)
```

```python
dist = norm(5, 0.5)
dist.pdf(0.2), dist.cdf(0.2)
```

```python
plot_gaussian_cdf(0, 1)
```

## Let's Code: How to estimate a probabiliity distribution?

### Parametric Estimation

Parametric estimation is the process of estimating the parameters of a probability distribution. For example, if we have a dataset of heights, we can estimate the parameters of a normal distribution by using the mean and standard deviation of the dataset.

### Non-Parametric Estimation

Non-parametric estimation is the process of estimating the parameters of a probability distribution without making any assumptions about the distribution. For example, if we have a dataset of heights, we can estimate the parameters of a normal distribution by using the mean and standard deviation of the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
from scipy.stats import norm
```

```python
sample = normal(size=1000)
plt.hist(sample, bins=30)
plt.title('Normal Distribution Simulation')
plt.show()
```

```python
# Parametric Estimation

sample = normal(loc=50, scale=5, size=1000) # mu = 50, sigma = 5
mu = np.mean(sample)
sigma = np.std(sample)
dist = norm(mu, sigma)
value = [i for i in range(30, 70)]
probabilities = [dist.pdf(i) for i in value]
plt.hist(sample, bins=30, density=True)
plt.plot(value, probabilities)
plt.title('Normal Distribution Simulation')
plt.show()
```

```python
# Nonparametric Estimation

from numpy import hstack
from sklearn.neighbors import KernelDensity

sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = hstack((sample1, sample2))
```

```python
model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)
```

```python
values = np.asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)

plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.title('Normal Distribution Simulation')
plt.show()
```

## What is MLE?

MLE stands for Maximum Likelihood Estimation. It is a method of estimating the parameters of a probability distribution by maximizing a likelihood function. For example, if we have a dataset of heights, we can estimate the parameters of a normal distribution by using the mean and standard deviation of the dataset.

The steps to calculate the MLE are:

1. Choose a probability distribution.

2. Choose the parameters of the probability distribution.

Then, MLE is an optimization problem where we want to find the parameters of the probability distribution that maximize the likelihood function.

$ P(X; \theta) = L(X; \theta) $

$ max L(X; \theta) -> max \prod P(X_i; \theta) $

$ max \log L(X; \theta) -> max \sum \log P(X_i; \theta) $

Where:

- $ P(X; \theta) $ is the probability of the dataset X given the parameters $ \theta $.

- $ L(X; \theta) $ is the likelihood function.

- $ X_i $ is the ith data point in the dataset X.

- $ \theta $ is the parameters of the probability distribution.

## Let's Code: MLE as base for Logistic Regression

Let's consider the MLE problem:

$ max \sum_i \log P(y_i|x_i; h) $

Where:

- $ y_i $ is the ith label in the dataset.

- $ x_i $ is the ith data point in the dataset, and each data point satisfies a Bernoulli distribution:

$ P(x_i) = \begin{cases} p & \text{if } y = 1 \\ 1 - p & \text{if } y = 0 \end{cases} $

In this case the likelihood function is:

$ L = \hat{y}y + (1 - \hat{y})(1 - y) $

This function has a maximum at $ \hat{y} ~ y $

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd
```

```python
def likelihood(y, yp):
    return yp * y + (1 - yp) * (1 - y)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

Y = np.arange(0, 1, 0.01)
YP = np.arange(0, 1, 0.01)
Y, YP = np.meshgrid(Y, YP)
Z = likelihood(Y, YP)

surf = ax.plot_surface(Y, YP, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

Considering $ p -> \log(p) $,and summing the likelihood function for all data points, we get:

$ max \sum_i (y \log \hat{y} + (1 - y) \log (1 - \hat{y})) $

$ = min - \sum_i (y \log \hat{y} + (1 - y) \log (1 - \hat{y})) $

Which is the same as the cross entropy function:

$ - \sum_i y_i \log \hat{y_i} $

## Bayes' Theorem

Bayes' Theorem is a formula that describes how to update the probabilities of hypotheses when given evidence. It follows simply from the axioms of conditional probability, but can be used to powerfully reason about a wide range of problems involving belief updates.

$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $

Where:

- $ P(A|B) $ is the probability of A given B.

- $ P(B|A) $ is the probability of B given A.

- $ P(A) $ is the probability of A.

- $ P(B) $ is the probability of B.

## Probability Examples

### Example 1: Dices

Let's say we have two dices. The first dice is a fair dice and the second dice is a loaded dice. The loaded dice has a higher probability of rolling a 6 than the fair dice. Let's say we roll both dices and we get a 6 on both dices. What is the probability that the first dice is the loaded dice?

We can use Bayes' Theorem to solve this problem. Bayes' Theorem is:

$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $

The solution to this problem is:

$ P(loaded|6) = \frac{P(6|loaded) \times P(loaded)}{P(6)} = \frac{1 \times \frac{1}{2}}{\frac{1}{6} \times \frac{1}{2} + 1 \times \frac{1}{2}} = \frac{6}{7} $

### Example 2: Cancer

Let's say we have a test that can detect cancer. The test is 99% accurate, meaning that the test will detect cancer 99% of the time when cancer is present. The test is also 99% accurate, meaning that the test will not detect cancer 99% of the time when cancer is not present. Let's say we take the test and the test comes back positive. What is the probability that we have cancer?

We can use Bayes' Theorem to solve this problem. Bayes' Theorem is:

$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $

The solution to this problem is:

$ P(cancer|positive) = \frac{P(positive|cancer) \times P(cancer)}{P(positive)} = \frac{0.99 \times 0.01}{0.99 \times 0.01 + 0.01 \times 0.99} = 0.5 $

### Example 3: Monty Hall

Let's say we are on a game show and we are given the choice of three doors. Behind one door is a car and behind the other two doors are goats. We pick a door, say door 1, and the host, who knows what's behind the doors, opens another door, say door 3, which has a goat. The host then asks us if we want to switch our choice to door 2. What should we do?

We can use Bayes' Theorem to solve this problem. Bayes' Theorem is:

$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $

The solution to this problem is:

$ P(car|door 3) = \frac{P(door 3|car) \times P(car)}{P(door 3)} = \frac{0.5 \times 0.33}{0.5 \times 0.33 + 1 \times 0.33} = 0.33 $

### What is MAP?

MAP stands for Maximum A Posteriori. It is a method of estimating the parameters of a probability distribution by maximizing a posterior probability. For example, if we have a dataset of heights, we can estimate the parameters of a normal distribution by using the mean and standard deviation of the dataset.

The steps to calculate the MAP are:

1. Choose a probability distribution.

2. Choose the parameters of the probability distribution.

Then, MAP is an optimization problem where we want to find the parameters of the probability distribution that maximize the posterior probability.

$ P(h|D) = \frac{P(D|h) \times P(h)}{P(D)} $

$ max P(h|D) -> max P(D|h) \times P(h) $

## What is Naive Bayes?

Naive Bayes is a classification algorithm that is based on Bayes' Theorem. It is called naive because it assumes that the features are independent of each other. For example, if we have a dataset of emails and we want to classify them as spam or not spam, we can use naive bayes to determine the likelihood of an email being spam given some words in the email.

$ P(y|x_1, x_2, ..., x_n) = \frac{P(x_1, x_2, ..., x_n|y) \times P(y)}{P(x_1, x_2, ..., x_n)} $
